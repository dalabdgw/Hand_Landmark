```python
import cv2
from google.colab.patches import cv2_imshow
import math
import numpy as np
import pandas as pd ## shin hyeong

DESIRED_HEIGHT = 480
DESIRED_WIDTH = 480
def resize_and_show(image):
  h, w = image.shape[:2]
  if h < w:
    img = cv2.resize(image, (DESIRED_WIDTH, math.floor(h/(w/DESIRED_WIDTH))))
  else:
    img = cv2.resize(image, (math.floor(w/(h/DESIRED_HEIGHT)), DESIRED_HEIGHT))
  cv2_imshow(img)

# Read images with OpenCV.
images = {name: cv2.imread(name) for name in uploaded.keys()}
# Preview the images.
#for name, image in images.items():
#  print(name)   
#  resize_and_show(image)
```

```python
data1_tubuh = {}
data_left = {}
data_right = {}

data_half = int(len(hand_tubuh)/2) 

for i in range(len(hand_tubuh)):
  if (i<data_half): data_left.update({hand_tubuh[i] : 0 })
  else: data_right.update({hand_tubuh[i]:0})
  data1_tubuh.update({
      hand_tubuh[i] : 0
  })

with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.7) as hands:
  for name, image in images.items():
    # Convert the BGR image to RGB, flip the image around y-axis for correct 
    # handedness output and process it with MediaPipe Hands.
    results = hands.process(cv2.flip(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), 1))

    # Print handedness (left v.s. right hand).
    ##print(f'Handedness of {name}:')
    ##print(results.multi_handedness)

    if not results.multi_hand_landmarks:
      cv2.imwrite('output_'+name, cv2.flip(image, 1)); ## shin hyeong
      alldata.append(data1_tubuh) 
      continue

    # Draw hand landmarks of each hand.
    print(f'Hand landmarks of {name}:')
    image_hight, image_width, _ = image.shape
    annotated_image = cv2.flip(image.copy(), 1)
    #for hand_landmarks in results.multi_hand_landmarks:

    flag = True
    flag2 = False
    data4_tubuh = {}
    if (len(results.multi_hand_landmarks)==1):
      y = str(results.multi_handedness[0]).split(':')[3].split()[0]
      flag2 = True
      if (y=="\"Right\""): data4_tubuh.update(data_left)
      else: flag = False

    data3_tubuh = {}
    count=0
    for hand_landmarks in results.multi_hand_landmarks:
      # Print index finger tip coordinates.
      ##print(
      ##    f'Index finger tip coordinate: (',
      ##    f'{hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x * image_width}, '
      ##    f'{hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y * c})'
      ##)
      mp_drawing.draw_landmarks(
          annotated_image,
          hand_landmarks,
          mp_hands.HAND_CONNECTIONS,
          mp_drawing_styles.get_default_hand_landmarks_style(),
          mp_drawing_styles.get_default_hand_connections_style())
      
      for i in range(len(hand_landmarks.landmark)):
        if (flag2): ## 1개일 경우
          if (flag): ## 오른쪽
            data4_tubuh.update({hand_tubuh[data_half+3*i]:hand_landmarks.landmark[i].x*image_width })
            data4_tubuh.update({hand_tubuh[data_half+3*i+1]:hand_landmarks.landmark[i].y*image_hight })
            data4_tubuh.update({hand_tubuh[data_half+3*i+2]:hand_landmarks.landmark[i].z })
          else:
            data4_tubuh.update({hand_tubuh[3*i]:hand_landmarks.landmark[i].x*image_width })
            data4_tubuh.update({hand_tubuh[3*i+1]:hand_landmarks.landmark[i].y*image_hight })
            data4_tubuh.update({hand_tubuh[3*i+2]:hand_landmarks.landmark[i].z })
        else:
          data3_tubuh.update({hand_tubuh[data_half*count+3*i]:hand_landmarks.landmark[i].x*image_width })
          data3_tubuh.update({hand_tubuh[data_half*count+3*i+1]:hand_landmarks.landmark[i].y*image_hight })
          data3_tubuh.update({hand_tubuh[data_half*count+3*i+2]:hand_landmarks.landmark[i].z })
      count+=1

    resize_and_show(cv2.flip(annotated_image, 1))
    if (flag2==False): alldata.append(data3_tubuh)
    cv2.imwrite('output_'+name, cv2.flip(annotated_image, 1)); ## shin hyeong

    if (flag == False): data4_tubuh.update(data_right)
    if (flag2): alldata.append(data4_tubuh)

df = pd.DataFrame(alldata)
df.to_csv('hand_result.csv')
```

```python
# Run MediaPipe Hands and plot 3d hands world landmarks.
with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.7) as hands:
  for name, image in images.items():
    # Convert the BGR image to RGB and process it with MediaPipe Hands.
    results = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    # Draw hand world landmarks.
    print(f'Hand world landmarks of {name}:')
    if not results.multi_hand_world_landmarks:
      continue
    for hand_world_landmarks in results.multi_hand_world_landmarks:
      mp_drawing.plot_landmarks(
        hand_world_landmarks, mp_hands.HAND_CONNECTIONS, azimuth=5)
```
